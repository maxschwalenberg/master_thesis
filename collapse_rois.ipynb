```python
from utils.config import Configuration, load_config
config = load_config("config.yaml")


value_to_set = 20
```

```python
import os
import numpy as np
import nibabel as nib


subj = 1
source_dir = config.t_test_roi_dir

lh_path = os.path.join(source_dir, f"lh.subj{subj:02d}.testrois.mgz")
rh_path = os.path.join(source_dir, f"rh.subj{subj:02d}.testrois.mgz")

new_lh_out = os.path.join(source_dir, f"lh.subj{subj:02d}.collapsed.testrois.mgz")
new_rh_out = os.path.join(source_dir, f"rh.subj{subj:02d}.collapsed.testrois.mgz")


import nibabel as nib

# Load MGZ files
lh_img = nib.load(lh_path)
rh_img = nib.load(rh_path)

# Get the data arrays
lh = lh_img.get_fdata()
rh = rh_img.get_fdata()

# Modify the data
lh[lh >= 20] = value_to_set
rh[rh >= 20] = value_to_set

# Create new nifti images with the modified data
new_lh_img = nib.Nifti1Image(lh, lh_img.affine, lh_img.header)
new_rh_img = nib.Nifti1Image(rh, rh_img.affine, rh_img.header)

# Save the new images
nib.save(new_lh_img, new_lh_out)
nib.save(new_rh_img, new_rh_out)


```
