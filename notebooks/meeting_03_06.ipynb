```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Read in the results
df_results = pd.read_excel("data/mantel_results_face_locations_2.xlsx")

# Compute the absolute maximum r across all data for symmetric color scale
max_abs_r = df_results['r_obs'].abs().max()

# Loop over each feature and plot r_obs and p_value side by side
for feature in df_results['feature'].unique():
    # Pivot tables for this feature
    heatmap_r = (
        df_results[df_results['feature'] == feature]
        .pivot(index='subject', columns='mask', values='r_obs')
    )
    heatmap_p = (
        df_results[df_results['feature'] == feature]
        .pivot(index='subject', columns='mask', values='p_value')
    )
    
    # Create one figure with two subplots (1 row, 2 columns)
    fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)
    
    # Left: r_obs heatmap
    sns.heatmap(
        heatmap_r,
        ax=axes[0],
        annot=True,
        fmt=".2f",
        cmap="coolwarm",
        center=0,
        vmin=-max_abs_r,
        vmax=max_abs_r,
        linewidths=0.5,
        linecolor='gray',
        mask=heatmap_r.isna(),
        cbar_kws={'label': 'Effect Size (r_obs)'}
    )
    axes[0].set_title(f"RSA Effect Sizes (Mantel r)\nFeature: {feature}")
    axes[0].set_xlabel("Mask (ROI)")
    axes[0].set_ylabel("Subject")
    
    # Right: p-value heatmap
    sns.heatmap(
        heatmap_p,
        ax=axes[1],
        annot=True,
        fmt=".3f",
        cmap="YlGnBu",
        mask=heatmap_p.isna(),
        cbar_kws={'label': 'Permutation p-value'}
    )
    axes[1].set_title(f"Permutation p-values\nFeature: {feature}")
    axes[1].set_xlabel("Mask (ROI)")
    # We share the y-axis labels, so no need to set ylabel again
    
    plt.tight_layout()
    plt.show()
```
