```python
import numpy as np
from scipy import stats

# Different sample sizes

s1 = [2]*5
s2 = [0,1,2,3,4]


group1 = np.random.normal(loc=12, scale=0.5, size=1000)   # 50 samples
group2 = np.random.normal(loc=12, scale=0.5, size=1000)  # 100 samples

t_statistic, p_value = stats.ttest_ind(s1, s2, equal_var=False)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```
